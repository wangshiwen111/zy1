```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 加载数据
data = pd.read_csv('US-pumpkins.csv')

# 查看数据的基本信息
print(data.info())
print(data.head())

# 选择用于聚类的特征
# 假设我们选择以下特征：Low Price, High Price, Item Size, Color
features = ['Low Price', 'High Price', 'Item Size', 'Color']

# 将非数值特征进行编码
label_encoder = LabelEncoder()
data['Item Size'] = label_encoder.fit_transform(data['Item Size'])
data['Color'] = label_encoder.fit_transform(data['Color'])

# 选择特征列
X = data[features]

# 标准化特征
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 使用肘部法则确定最佳的聚类数量
inertia = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# 绘制肘部图
plt.figure(figsize=(10, 6))
plt.plot(range(2, 10), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()

```

```python
# 根据肘部图选择最佳聚类数量
# 假设选择 k=3
kmeans = KMeans(n_clusters=6, random_state=42)
kmeans.fit(X_scaled)
kmeans_labels = kmeans.labels_

# 计算轮廓系数
silhouette_avg = silhouette_score(X_scaled, kmeans_labels)
print(f'Silhouette Score for K-Means: {silhouette_avg}')
```

```python
# 使用DBSCAN进行聚类
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X_scaled)
dbscan_labels = dbscan.labels_

# 计算轮廓系数
dbscan_labels_filtered = dbscan_labels[dbscan_labels != -1]
X_scaled_filtered = X_scaled[dbscan_labels != -1]

silhouette_avg = silhouette_score(X_scaled_filtered, dbscan_labels_filtered)
print(f'Silhouette Score for DBSCAN: {silhouette_avg}')
```

```python
# 将聚类结果添加到原始数据中
data['KMeans_Cluster'] = kmeans_labels
data['DBSCAN_Cluster'] = dbscan_labels

# 可视化K-Means聚类结果
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Low Price', y='High Price', hue='KMeans_Cluster', data=data, palette='viridis')
plt.title('K-Means Clustering')
plt.show()

# 可视化DBSCAN聚类结果
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Low Price', y='High Price', hue='DBSCAN_Cluster', data=data, palette='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```
